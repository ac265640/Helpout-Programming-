```python
import matplotlib.pyplot as plt 
import numpy as np

#AVERAGE POWERPLAY SCORE (2022)

x=np.array(["CSK", "DC", "GT", "KKR", "LSG", "MI", "PBSK", "RR","RCB", "SRH"])

y=np.array([40.9,57.2,47.0,41.6,42.7,45.8,53.7,48.6,39.5,42.8]) 
mycolors=["#FFD700", "b", "#000088", "k", "#32CD32", "#6495ED", "#DC143C", "#EE1289", "#CD3333","#FF6103"] 
plt.bar(x,y,color=mycolors,width=0.5)

plt.title("AVERAGE POWERPLAY SCORE (2022)")

plt.xlabel("TEAM")

plt.ylabel("RUNS")

plt.show()

#AVERAGE WICKETS LOST (2022)

x=np.array(["CSK", "DC", "GT", "KKR", "LSG", "MI", "PBSK", "RR","RCB", "SRH"])

y=np.array([1.8,1.4,1.3,2.0,1.6,1.3,1.6,1.2,1.7,1.3])

mycolors=["#FFD700", "b", "#000088", "k", "#32CD32", "#6495ED", "#DC143C", "#EF1289", "#CD3333","#FF6103"]

plt.bar(x,y,color=mycolors,width=0.5) 
plt.title("AVERAGE WICKETS LOST(2022)")

plt.xlabel("TEAM")

plt.ylabel("RUNS")

plt.show()
```
